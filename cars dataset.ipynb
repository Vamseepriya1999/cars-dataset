```python
import numpy as np
import pandas as pd
```

```python
cars=pd.read_csv('../input/carsdata/cars.csv')
```

```python
cars.describe()
```

```python
cars.columns
```

```python
cars.head()
```

```python
y=cars.mpg
```

```python
X=cars[[' cylinders',' cubicinches',' hp', ' weightlbs',]]
```

```python
X.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,
                                                      random_state=0)
```

```python
X_train.head()
```

```python
#categorical variables
s = (X_train.dtypes == 'object')
object_cols = list(s[s].index)

print("Categorical variables:")
print(object_cols)
```

```python
#one hot encoding to convert categorical variables to numerical 
from sklearn.preprocessing import OneHotEncoder
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols_train = pd.DataFrame(OH_encoder.fit_transform(X_train[object_cols]))
OH_cols_valid = pd.DataFrame(OH_encoder.transform(X_valid[object_cols]))


OH_cols_train.index = X_train.index
OH_cols_valid.index = X_valid.index


num_X_train = X_train.drop(object_cols, axis=1)
num_X_valid = X_valid.drop(object_cols, axis=1)

OH_X_train = pd.concat([num_X_train, OH_cols_train], axis=1)
OH_X_valid = pd.concat([num_X_valid, OH_cols_valid], axis=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
cars_model=DecisionTreeRegressor(random_state=0)
cars_model.fit(OH_X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error
predictions=cars_model.predict(OH_X_valid)
print("MEA:")
print(mean_absolute_error(y_valid,predictions))
```

```python
from sklearn.ensemble import RandomForestRegressor
cars_model1=RandomForestRegressor(random_state=1)
cars_model1.fit(OH_X_train,y_train)
```

```python
predictions1=cars_model1.predict(OH_X_valid)
print("MEA:")
print(mean_absolute_error(y_valid,predictions1))
```

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(cars_model1,OH_X_train,y_train,cv=5)
print("score:",score.mean())
```

```python
from matplotlib import pyplot as plt
plt.title('score')
plt.xlabel('predictions')
plt.ylabel('y_valid')
plt.plot(predictions1,y_valid)
plt.show()
```
